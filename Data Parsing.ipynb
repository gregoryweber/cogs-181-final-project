```python
def oldyoung(x):
    if x == "Younger adult":
        return "Young"
    else:
        return "Old"

HAI['Subject Phenotype'] = HAI['Subject Phenotype'].apply(oldyoung)
HAI.rename(columns={'Subject Phenotype': 'Age Group', 'Study Time Collected': 'Day'}, inplace=True)
HAI.head()
```

```python
#So day is slightly different, but otherwise probably ok to assume. HAI titer still later so probably ok. We will go based on HAI Titer Document with actual HAI data.
HAI1.drop(columns = ['Subject','Age Group'], inplace=True)
HAI.drop(columns = 'Day', inplace=True)
HAIall = HAI.join(HAI1)
HAIall.tail()
```

```python
HAIall = HAIall.loc[HAIall['Age Group'] == 'Young']
HAIall = HAIall.loc[HAIall['Day'] == 28]
display(HAIall)
HAIall.shape
```

```python
HAIall.drop(columns = ['Age Group','Planned Visit Accession','Day'], inplace = True)
HAIall.reset_index(drop=True, inplace = True)
HAIall.head()
#Y values: H1, H3, B
```

```python
All = pd.merge(HAIall, FCS[['Subject Accession','File Name']], on = 'Subject Accession')
All.rename(columns = {'File Name':'FCS Raw'}, inplace = True)
display(All)
```

```python
# For Verification of FCS File Mapping -- done manually
All = pd.merge(All, WSP[['Subject Accession','File Name']], on = 'Subject Accession')
All.rename(columns = {'File Name':'WSP File'}, inplace = True)
display(All)
```

```python
# Add Preprocessing File Names 
def preprocessfile(name):
    names = name.split('.')
    return "export_" + names[0] + "_Viability." + names[2]

All['FCS Preprocessed'] = All['FCS Raw'].apply(preprocessfile)
All = All.iloc[:,[0, 1, 2, 3, 4, 6, 5]]
display(All)
```

## CyTOF Data Processing

```python
meta = pd.read_csv('Metadata/SDY112-DR54_Subject_2_CyTOF_result.txt', sep="\t")
meta = meta[['Subject Accession','ARM Accession','Planned Visit Accession','File Name']]
meta.head()
```

```python
meta = meta[meta['ARM Accession'] == 'ARM642']
meta = meta[meta['Planned Visit Accession'] == 'PV1950']
meta = meta[meta['File Name'].str.contains('fcs')]
meta = meta[['Subject Accession','File Name']]
meta
```

```python
hai = pd.read_csv('HAI/hai_result.txt', sep="\t")
hai.head()
```

```python
hai = hai[hai['ARM_ACCESSION'] == 'ARM642']
hai = hai[hai['STUDY_TIME_COLLECTED'] == 28]
display(hai.head())
hai['VALUE_REPORTED'].unique()
```

```python
hai = hai[['SUBJECT_ACCESSION','VALUE_REPORTED','VIRUS_STRAIN_REPORTED']]
hai.columns = ['Subject Accession', 'Value','Virus']
hai.head()
```

```python
# a check
hai.merge(meta, on = 'Subject Accession')['Subject Accession'].value_counts()
```

```python
# well that's weird -- dropping sample
hai[hai['Subject Accession'] == 'SUB116537']
```

```python
hai.drop(hai[hai['Subject Accession'] == 'SUB116537'].index, inplace=True)
meta.drop(meta[meta['Subject Accession'] == 'SUB116537'].index, inplace=True)
alld = meta.copy()
alld
```

```python
def findTiter(subject, virus):
    df = hai.loc[hai['Subject Accession'] == subject]
    df = df.loc[df['Virus'] == virus]
    return df['Value'].item()

alld['H1'] = alld['Subject Accession'].apply(findTiter,args=["A/California/07/2009"])
alld['H3'] = alld['Subject Accession'].apply(findTiter,args=["A/Perth/16/2009"])
alld['B'] = alld['Subject Accession'].apply(findTiter,args=["B/Brisbane/60/2008"])
alld
```

```python
import os
import bokeh
from bokeh.plotting import show
import pandas as pd

import flowkit as fk

bokeh.io.output_notebook()
```

```python
workspace = fk.Workspace('Metadata/Study 15-2011-V1.504090.wsp', fcs_samples='FCS/CyTOF/Raw', ignore_missing_files = True)
```

```python
samples = workspace.get_sample_ids()
```

```python
workspace.analyze_samples()
```

```python
workspace.get_gate_events('082812-Mike-Study 15-2011-15065_cells_found.txt', gate_name = 'Live intact singlets', source = "raw").index
```

```python
for s in samples:
    idx = workspace.get_gate_events(s, gate_name = 'Live intact singlets', source = "raw").index
    flag = np.arange(len(workspace.get_gate_events(s, source = "raw")))
    mask=np.full(len(flag),True,dtype=bool)
    mask[idx]=False
    flag = flag[mask]
    sample = workspace.get_sample(s)
    sample.set_flagged_events(flag)
    name = s.split('.')[0] + '_preprocessed.fcs'
    sample.export(name, source='raw', exclude_flagged=True, directory="FCS/CyTOF/Preprocessed")
```
